```julia
#Tarea Analisis del ruido en senales

#Objetivo:
#En el presente ejercicio, se pretende demostrar como es el cambio de una senal, cuando se le agrega un ruido random
#para dichos fines, se graficaran una senal con ruido y otra sin ruido, observandose las diferencias y discutiendose posteriormente 
```

```julia
#primeramente se generan los valores de entrada que permitiran graficar una funcion senosoidal sin ruido

f1=2 #frec 1
f2=3 #frec 2
seg=10 #tiempo de muestreo
sps=8 #muestras por segundo
t=collect(linspace(0,seg,seg*sps)) #vector de tiempo

#la funcion sin ruido viene dada por
y=cos.(2*π*t*f1) + sin.(2*π*t*f2);

#La funcion con ruido sera expresada de la siguiente manera y2=cos(f1)+sin(f2)+ruido
#y como se dijo anteriormente el ruido tiene que ser aleatorio, por lo que

noise=rand(length(y)); 

y2=cos.(2*π*t*f1) + sin.(2*π*t*f2)+ noise;
```

```julia
using PyPlot
```

```julia
subplot(2,1,1)
plot(y)
title("senal sin ruido")
subplot(2,1,2)
plot(y2)
title("senal con ruido")
xlabel("frecuencia")
ylabel("amplitud")
## Se puede apreciar que si bien los altos y bajos de las funciones, coinciden, la amplitud de estos varia a
#lo largo del espectrograma 
```

```julia
#se calculan las transformadas de fourier para ambas senales, tanto para la senal con ruido como la que no tiene ruido
f=abs.(fft(y,1));
fn=abs.(fft(y2,1));
```

```julia
#se construyen los vectores de entrada para el grafico de las transformadas de fourier

#funcion sin ruido
frecvec=collect(linspace(0,sps,sps*seg));
length(frecvec);
fr=length(f);
fr2=0.5*fr
fr2=convert(Int32,fr2)

# funcion con ruido

frecvec=collect(linspace(0,sps,sps*seg));
length(frecvec);
fr3=length(f);
fr4=0.5*fr3
fr4=convert(Int32,fr4)
```

```julia
#senal sin ruido
subplot(2,1,1)
plot(frecvec[1:fr2],f[1:fr2],"k")
title("FFT SENAL SENOSOIDAL")
xlabel("Frec [Hz]")
ylabel("Amplitud")

subplot(2,1,2)
plot(t,y)
title("senal en tiempo")
xlabel("Tiempo")
ylabel("Amplitud")
```
